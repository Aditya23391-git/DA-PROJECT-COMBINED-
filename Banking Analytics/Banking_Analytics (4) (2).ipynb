```python
#Examine the distribution of unique categorical columns 
categorical_cols=df[["BRId","GenderId","Amount of Credit Cards","Nationality","Fee Structure","Occupation","Loyalty Classification","Properties Owned","Risk Weighting","IAId","Income Band"]].columns
for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

## UNIVARIATE ANALYSIS

```python
for i, predictor in enumerate(df[[
    "BRId", "GenderId","Amount of Credit Cards",
    "Nationality",
    "Fee Structure",
   
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "IAId",
    "Income Band"]].columns):

    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x=predictor)
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {predictor}")
    plt.show()
```

## Bivariate Analysis

```python
for i, predictor in enumerate(df[[
    "BRId", "GenderId","Amount of Credit Cards",
    "Nationality",
    "Fee Structure",
   
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "IAId",
    "Income Band"]].columns):

    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x=predictor,hue='Nationality')
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {predictor}")
    plt.show()
```

```python
# Histogrram of different occupation
for col in categorical_cols :
    if col == "Occupation":
        continue
    plt.figure(figsize=(7,4))
    sns.histplot(df[col])
    plt.title("Histogram of occupation count")
    plt.ylabel("Count")
    plt.xlabel(col)
    plt.show()
```

# Numerical Analysis

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
#Univariate Analysis and visualization

for i, col in enumerate(numerical_cols):
    plt.figure(figsize=(8,4))
   ## plt.subplot(2,1,i+1) : we can use this too instead of plt.figure
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

## HEATMAP

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
correlation_matrix=df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix ,annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

## INSIGHTS

```python
## The correlation heatmap shows a strong positive relationship between Checking Accounts, Savings Accounts, and Bank Deposits. This indicates that customers maintaining higher balances in their checking and savings accounts also tend to hold larger overall bank deposits.
## These customers may represent financially stable and high-value banking clients.
```
