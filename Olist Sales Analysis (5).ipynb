## IMPORTING DATASET

```python
#We use this when files are inside folder
import os
# Folder path 
path = r"C:\Users\Aditya Singh\Downloads\archive (4)"

# Load datasets
orders = pd.read_csv(os.path.join(path, "olist_orders_dataset.csv"))
order_items = pd.read_csv(os.path.join(path, "olist_order_items_dataset.csv"))
products = pd.read_csv(os.path.join(path, "olist_products_dataset.csv"))
customers = pd.read_csv(os.path.join(path, "olist_customers_dataset.csv"))
reviews = pd.read_csv(os.path.join(path, "olist_order_reviews_dataset.csv"))
category_translation = pd.read_csv(os.path.join(path, "product_category_name_translation.csv"))

print("Files loaded successfully")
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

orders['order_year'] = orders['order_purchase_timestamp'].dt.year
orders['order_month'] = orders['order_purchase_timestamp'].dt.strftime('%Y-%m')
```

```python
## FILLING NULL VALUES WITH N/A
products['product_category_name'] = products['product_category_name'].fillna('N/A')
```

```python
## SELECTING COLUMNS AND REMOVING UNECESSARY ONES
products = products[['product_id', 'product_category_name']]
```

```python
order_items.info()
```

```python
## FINDING REVENUE
order_items['revenue']=order_items['freight_value']+order_items['price']
```

```python
customers.shape
customers.info()
```

```python
customers.isnull().sum()
customers.duplicated().sum()
```

```python
## SELECTING COLUMNS AND REMOVING UNECESSARY ONES
customers = customers[
    ['customer_id', 'customer_city', 'customer_state']
]
```

```python
customers.head()
```

```python
category_translation.shape
category_translation.info()
```

```python
## CHECKING SUM OF DUPLICATES
category_translation.duplicated().sum()
```

```python
## SELECTING COLUMNS AND REMOVING UNECESSARY ONES
reviews = reviews[['order_id', 'review_score']]
```

```python
reviews.head()
```

```python
reviews['review_score'].value_counts().sort_index()
```

```python
## MERGING ALL CLEANED DATASETS INTO FINAL DATASET
df=pd.merge(orders,order_items ,on='order_id',how='left')
df=pd.merge(df,products , on ='product_id',how='left')
df = pd.merge(df,category_translation,on='product_category_name',how='left')
df = pd.merge(df,customers,on='customer_id',how='left')
final_df = pd.merge(df,reviews,on='order_id',how='left')
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
## DROPPING UNCECCESARY COLUMNS
final_df.drop(columns=['product_category_name'], inplace=True)
```

```python
## VIEWING COLUMNS IN FINAL DATASET
final_df.columns
```

```python
final_df['delivery_days'] = (
    final_df['order_delivered_customer_date']
    - final_df['order_purchase_timestamp']
).dt.days
```

```python
final_df.head()
```

```python
## SORTING VALUES IN ASCENDING ORDER
final_df.isnull().sum().sort_values(ascending=False)
```

```python
## FILLING NULL VALUES WITH UNKNOWN
final_df['product_category_name_english'] = (
    final_df['product_category_name_english']
    .fillna('Unknown')
)
```

```python
## ROWS WITH MISSING VALUES ARE DELETED
final_df = final_df[final_df['revenue'].notna()]
```

```python
final_df.isnull().sum().sort_values(ascending=False)
```

```python
## EXPORTING FINAL DATASET
output_path =r"C:\Users\Aditya Singh\Downloads\archive (4)"
os.makedirs(output_path ,exist_ok=True)
final_df.to_csv(os.path.join(output_path,"olist_cleaned.csv"),index=False)
print("File Cleaned")
```
